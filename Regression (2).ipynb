```python
#Cole Doherty and Chris Beatrez
import pandas as pd
import statistics as st
#import mat
import matplotlib.pyplot as plt
import seaborn as sns
import numpy
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)

#Reads in the csv file into a dataframe
df = pd.read_csv('PlantHeight.csv')

#Slope calculations for Temperature/Height
def calcTH(x,y):
    calculate = 0
    div= 0
    slope = 0
    yInt = 0
 
    avCx = st.mean(x)
    avCy = st.mean(y)

    for i in range(len(x)):
       calculate += ((x[i]-avCx)*(y[i]-avCy))
       div += (pow((x[i]-avCx),2))

    slope = calculate/div
    print("slope (b) = ",slope)
    yInt = (avCy-(slope*avCx))
    print("Y-int = ",yInt)
    print("y =(",slope,"* x) + ",yInt," \n")
    
#Slope calculations for Rain/Height   
def calcHR(z,y):
    calculate = 0
    div= 0
    slope = 0
    yInt = 0
 
    avCz = st.mean(z)
    avCy = st.mean(y)

    for i in range(len(z)):
       calculate += ((z[i]-avCz)*(y[i]-avCy))
       div += (pow((z[i]-avCz),2))

    slope = calculate/div
    print("slope (b) = ",slope)
    yInt = (avCy-(slope*avCz))
    print("Y-int = ",yInt)
    print("y =(",slope,"* x) + ",yInt," \n")

#Slope calculations for Altitude/Height
def calcHA(b,y):
    calculate = 0
    div= 0
    slope = 0
    yInt = 0
 
    avCb = st.mean(b)
    avCy = st.mean(y)

    for i in range(len(b)):
       calculate += ((b[i]-avCb)*(y[i]-avCy))
       div += (pow((b[i]-avCb),2))

    slope = calculate/div
    print("slope (b) = ",slope)
    yInt = (avCy-(slope*avCb))
    print("Y-int = ",yInt)
    print("y =(",slope,"* x) + ",yInt," \n")

#Slope calculations for Site Number/Height
def calcHS(c,y):
    calculate = 0
    div= 0
    slope = 0
    yInt = 0
 
    avCc = st.mean(c)
    avCy = st.mean(y)

    for i in range(len(c)):
       calculate += ((c[i]-avCc)*(y[i]-avCy))
       div += (pow((c[i]-avCc),2))

    slope = calculate/div
    print("slope (b) = ",slope)
    yInt = (avCy-(slope*avCc))
    print("Y-int = ",yInt)
    print("y =(",slope,"* x) + ",yInt," \n")
   
#Getting all the values in the data and storing it into a variable.  
x = df["temp"]
y = df["height"]
z = df["rain"]
b = df["alt"]
c = df["site"]

calcTH(x,y)
calcHR(z,y)
calcHA(b,y)
calcHS(c,y)

#This is where all of the scatterplots are printed out.
sns.regplot(x = x, y = y)
plt.plot(x, y, 'o')

plt.xticks(rotation = 25)
plt.xlabel('Temperature(C)')
plt.ylabel('Height')
plt.title('Plant Height', fontsize = 20)
plt.grid()
plt.show()

#r squared calculations using numpy library
corr_matrix1 = numpy.corrcoef(x, y)
corr1 = corr_matrix1[0,1]
R_sq1 = corr1**2

print("Rsquared =", R_sq1)
#----------------------------------------------------------
sns.regplot(z, y = y)
plt.plot(z, y, 'o')

plt.xticks(rotation = 25)
plt.xlabel('Rain')
plt.ylabel('Height')
plt.title('Plant Height', fontsize = 20)
plt.grid()
plt.show()

corr_matrix2 = numpy.corrcoef(z, y)
corr2 = corr_matrix2[0,1]
R_sq2 = corr2**2

print("Rsquared =", R_sq2)
#----------------------------------------------------------
sns.regplot(b, y = y)
plt.plot(b, y, 'o')

plt.xticks(rotation = 25)
plt.xlabel('Altitude')
plt.ylabel('Height')
plt.title('Plant Height', fontsize = 20)
plt.grid()
plt.show()

corr_matrix3 = numpy.corrcoef(b, y)
corr3 = corr_matrix3[0,1]
R_sq3 = corr3**2

print("Rsquared =", R_sq3)
#----------------------------------------------------------
sns.regplot(c, y = y)
plt.plot(c, y, 'o')

plt.xticks(rotation = 25)
plt.xlabel('Site Number')
plt.ylabel('Height')
plt.title('Plant Height', fontsize = 20)
plt.grid()
plt.show()

corr_matrix4 = numpy.corrcoef(c, y)
corr4 = corr_matrix4[0,1]
R_sq4 = corr4**2

print("Rsquared =", R_sq4)
```
